```python
from os import path
import os
from pydub import AudioSegment
from scipy.io import wavfile
```

```python
# convert all songs to wav files
def convert_files(AUDIO_DIR, OUTPUT_DIR):
    for path in [os.path.join(AUDIO_DIR, p) 
                     for p in os.listdir(AUDIO_DIR) 
                     if not (p.endswith('checksums') or p.endswith('.txt') or p.endswith('.DS_Store'))]:
            # make the top level directory if needed
            print(path)
            try:
                os.mkdir(os.path.join(OUTPUT_DIR, path.split('/')[-1]))
            except:
                pass
            
            # add all songs to the list
            AUDIO_PATHS = [os.path.join(path, track) for track in os.listdir(path)]

            for song in AUDIO_PATHS:
                dst = os.path.join(OUTPUT_DIR,song.split('/')[-2],song.split('/')[-1].split('.')[0]+'.wav')
                try:
                    sound = AudioSegment.from_mp3(song)
                    sound.export(dst, format="wav")
                except:
                    pass
```

```python
SMALL_AUDIO_DIR = 'data/fma_small'
SMALL_OUTPUT_DIR = 'data/fma_small_wav'
```

```python
# small subset
convert_files(SMALL_AUDIO_DIR, SMALL_OUTPUT_DIR)
```

```python
# medium subset
MEDIUM_AUDIO_DIR = 'data/fma_medium'
MEDIUM_OUTPUT_DIR = 'data/fma_medium_wav'
    
convert_files(MEDIUM_AUDIO_DIR, MEDIUM_OUTPUT_DIR)
```
